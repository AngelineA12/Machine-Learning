**6. NON-PARAMETRIC METHODS**


For our dataset, Mann-Whitney U test will be suitable, so we choose Mann-Whitney U test

In the below code, we are converting the dataset into a dataframe

```python
df = pd.DataFrame(df)
```

In the below code, we can infer that,
* Europe_newcases = df[df['WHO Region'] == 'Europe']['New cases']: This line filters the DataFrame df to only include the 'New cases' data for the 'Europe' region. It selects the 'New cases' column where the 'WHO Region' is 'Europe'.

* Africa_newcases = df[df['WHO Region'] == 'Africa']['New cases']: This line filters the DataFrame df to only include the 'New cases' data for the 'Africa' region. It selects the 'New cases' column where the 'WHO Region' is 'Africa'.

* mann_whitney_test = stats.mannwhitneyu(Europe_newcases, Africa_newcases): This line performs the Mann-Whitney U test using SciPy's mannwhitneyu() function. This non-parametric test is used to determine whether two independent samples have different distributions. It takes the two sets of data ('Europe_newcases' and 'Africa_newcases') as input and returns the test statistic and p-value.

* print("\nMann-Whitney U test result:"): This line prints a header indicating the start of the Mann-Whitney U test result.

* print("U Statistic: ", mann_whitney_test.statistic): This line prints the U statistic, which is the test statistic computed by the Mann-Whitney U test. It represents the degree of difference or similarity in the distributions of the two samples.

* print("p-value:", mann_whitney_test.pvalue): This line prints the p-value computed by the Mann-Whitney U test. The p-value indicates the probability of observing the given test statistic (or a more extreme value) under the null hypothesis that the distributions of the two samples are the same. A low p-value suggests that the null hypothesis should be rejected, indicating a significant difference between the distributions.

```python
Europe_newcases = df[df['WHO Region'] == 'Europe']['New cases']
Africa_newcases = df[df['WHO Region'] == 'Africa']['New cases']
mann_whitney_test = stats.mannwhitneyu(Europe_newcases,Africa_newcases)
print("\nMann-Whitney U test result:")
print("U Statistic: ", mann_whitney_test.statistic) #The degree of distribution between the variables/values(Europe_newcases,Africa_newcases)
print("p=value:",mann_whitney_test.pvalue)
```

The result of the above code says that,
* U Statistic (1688.5):
The U statistic is a rank-sum statistic calculated by the Mann-Whitney U test. It quantifies the degree of difference or similarity in the distributions of the two samples ('Europe_newcases' and 'Africa_newcases').
In this case, the U statistic is 1688.5. Higher values of U suggest greater differences in the distributions of the two samples.

* p-value (0.0244):The p-value is a measure of the strength of evidence against the null hypothesis. It indicates the probability of obtaining the observed results, or results more extreme, if the null hypothesis is true.
A low p-value suggests that the observed results are unlikely to occur under the null hypothesis, leading to rejection of the null hypothesis.
In this case, the p-value is approximately 0.0244, which is below conventional significance levels such as 0.05 or 0.01.

* Interpretation:Since the p-value (0.0244) is less than the significance level (e.g., 0.05), we reject the null hypothesis.
Therefore, we conclude that there is a statistically significant difference in the number of new cases between the 'Europe' and 'Africa' regions.

* In summary, the results of the Mann-Whitney U test provide evidence that the distributions of new cases differ significantly between the 'Europe' and 'Africa' regions.
